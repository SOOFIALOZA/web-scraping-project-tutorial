```python
from bs4 import BeautifulSoup as bs
import requests
import time
url = "https://4geeks.com/"
page = requests.get(url)
html_string = page.text
print(html_string)

bs(html_string, "html.parser")
soup = bs(html_string, "html.parser")
print (soup.prettify())
```
